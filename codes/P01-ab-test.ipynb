# 1. Imports

```python
import numpy as np
import pandas as pd

import scipy.stats as stats

import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

# 2. Reading the dataset

```python
data = pd.read_csv('../datasets/conversion_daily_sample.csv')
```

```python
data.head()
```

# Plotting the control and treatment distribution

```python
plt.figure(figsize = (10, 8))
sns.distplot(data.Conversion_A, color='blue', label = 'Group A');
sns.distplot(data.Conversion_B, color = 'red', label = 'Group B');
plt.xlabel('Conversion rate', fontsize = 14)
plt.xticks(np.arange(0.05,0.35, 0.05), fontsize = 12)
plt.xlim(0.05,0.30)

plt.ylabel('Number', fontsize = 14)
plt.yticks(np.arange(0,22, 2), fontsize = 12)
plt.ylim(0,20)

plt.legend(fontsize = 12);
```

```python
print('Mean conversion rate for group A is: {} \nMean conversion rate for group B is: {}'.format(
    round(data.Conversion_A.mean(), 2), round(data.Conversion_B.mean(), 2))
     )
```

# t-student test

```python
t_stat, p_stat = stats.ttest_ind(data.Conversion_B, data.Conversion_A)

print('t_stat is: {} \np_stat is: {}'.format(t_stat, p_stat))
```
